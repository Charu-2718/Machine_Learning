```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
x=dataset.iloc[:,3:5].values

print(x)
```

To find the number of clusters using elbow method

```python

from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  km= KMeans(n_clusters=i , init='k-means++' ,random_state=1)
  km.fit(x)
  wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("No. of clusters")
plt.ylabel("WCSS")
plt.show()
```

from elbow method we get no of cluster equal to 5

```python
km= KMeans(n_clusters=i , init='k-means++' ,random_state=1)
y_km=km.fit_predict(x)
```

```python
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=100,c='yellow',label='C1')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=100,c='blue',label='C2')
plt.scatter(x[y_km==2,0],x[y_km==2,1],s=100,c='green',label='C3')
plt.scatter(x[y_km==3,0],x[y_km==3,1],s=100,c='red',label='C4')
plt.scatter(x[y_km==4,0],x[y_km==4,1],s=100,c='black',label='C5')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=300,c='cyan',label='centroid')
plt.title("KMeans")
plt.xlabel("income")
plt.ylabel("spending")
plt.show()
```
